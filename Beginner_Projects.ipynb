```python
import random
wordBank = "peach lychee coconut banana melon cherry lemon orange"
wordBank = wordBank.split(' ')
word = random.choice(wordBank)
wordLength = len(word)
display = ["_"]*wordLength
curDisplay = " ".join(display)
realDisplay = "".join(display)
# print(word)
# wordLength
print(f"Begin Guessing {wordLength} letters")
# for i in word:
#   print ("_", end= ' ')

print(" ".join(display))
chances = len(word)

guess = input("guess a letter (full word is fruit) ").lower()

while chances >= 1:
  while len(guess) !=1 or not guess.isalpha():
    guess = input("Please enter a single letter").lower()
  for i, letter in enumerate(word):
    if letter == guess:
      display[i] = guess
      curDisplay = " ".join(display)
      realDisplay = "".join(display)
  if realDisplay == word:
    print(f"You won! The word is {word}")
    break
  elif guess in word:
    guess = input(f"{curDisplay}, guess next letter ").lower()
  else:
    guess = input(f"Try again {curDisplay}")
    chances -= 1

if chances == 0:
  print("No more chances")
```

```python
import random
suits = ['Heart','Diamond','Clubs','Spade']
ranks = ['2','3','4','5','6','7','8','9','10','Jack','Queen','King','Ace']
deck = [(rank,suit) for rank in ranks for suit in suits]
game = True

def card_value(card):
  rank, suit = card
  if rank in ['Jack','Queen','King']:
    return 10
  elif rank in ['Ace']:
    return 11
  else:
    return int(rank)

random.shuffle(deck)
player_card = [deck.pop(),deck.pop()]
dealer_card = [deck.pop(),deck.pop()]

player_score = sum(card_value(card) for card in player_card)

if player_score == 21:
  print("Player wins!")
  game = False


while game == True:

  player_score = sum(card_value(card) for card in player_card)
  print(player_card)
  print(player_score)

  if player_score > 21:
    print("You bust :(")
    game = False
    break

  action = input("Hit or stay?").lower().strip()
  if action in ("hit","h","y"):
    player_card.append(deck.pop())
  elif action in ("stay","s","n"):
    print("You stay")
    break

while game == True:
  print("Dealer's turn")
  dealer_score = sum(card_value(card) for card in dealer_card)
  print(dealer_score)
  if dealer_score == 21:
    print("Dealer wins")
    game = False
    break
  elif dealer_score < 17:
    dealer_card.append(deck.pop())
  elif 17 <= dealer_score <= 21:
    break
  elif dealer_score > 21:
    print("Dealer busts you win!")
    game = False
    break

while game == True:
  if player_score > dealer_score:
    print("You win!")
  elif player_score == dealer_score:
    print("You tie")
  elif player_score < dealer_score:
    print("You lose")
  game = False
  break
```

```python
import random
import matplotlib.pyplot as plt
```

```python
class Player:
  def __init__(self, name, starting_cash):
    self.name = name
    self.cash = starting_cash
    self.portfolio = {}

  def buy(self, stock, amount):
    if amount > self.cash:
      print("Not enough cash")
      return

    shares = round(amount/stock.price, 2)
    self.cash -= amount
    self.portfolio[stock.name] = self.portfolio.get(stock.name, 0) + shares

    print(f"Bought {shares} shares of {stock.name} for ${amount:.2f}")

  def view_portf(self):
    print(f"\n{self.name}'s Portfolio")
    if not self.portfolio:
      print("- Empty portfolio")
      return
    for stock_name, shares in self.portfolio.items():
      print(f" - {stock_name}:{shares} shares")
    print(f"Cash: ${self.cash:.2f}")

  def sell(self, stock, shares):

    owned = self.portfolio.get(stock.name, 0)
    if shares > owned:
      print(f"You only own {owned} shares of {stock.name}. Can't sell {shares}.")
      return
    proceeds = round(shares*stock.price, 2)
    self.cash += proceeds
    self.portfolio[stock.name] = owned - shares

    if self.portfolio[stock.name] == 0:
      del self.portfolio[stock.name]
    print(f"Sold {shares} shares of {stock.name} for ${proceeds:.2f}")

  def portfolio_value(self, stocks):
    total_value = self.cash
    for stock in stocks:
        shares = self.portfolio.get(stock.name, 0)
        total_value += shares * stock.price
    return round(total_value, 2)



class Stock:

  def __init__(self, name, ticker, industry, price):
    self.name = name
    self.ticker = ticker
    self.industry = industry
    self.price = price

class Event:
  def __init__(self, name, desc, effects):
      self.name = name
      self.desc = desc
      self.effects = effects

  def apply(self, stocks):
    print(f"\nEvent: {self.name}")
    for stock in stocks:
      if stock.industry in self.effects:
        low, high = self.effects[stock.industry]
        change_percent = random.uniform(low, high)
        change_amount = round(stock.price*change_percent, 2)
        stock.price = round(stock.price + change_amount, 2)

        direction = "▲" if change_percent > 0 else "▼"
        print(f" {direction} {stock.name} ({stock.industry}): {change_percent*100:+.2f}% -> ${stock.price}")

```
